Get the study instance ID of the CT scan (not the DX)

```python
from pathlib import Path
import os
import pandas as pd

project_folder = Path(os.getcwd()).parent
dataset_folder = project_folder / "lidc_idri"
metadata = project_folder / "data" / "metadata.csv"
df = pd.read_csv(metadata)

# for loop placeholder
row = df[(df["Patient ID"] == "LIDC-IDRI-0001") & (df["Modality"] == "CT")].iloc[0]

# for loop should start here
# actual thing: for index, row in df.iterrows():
# if row["Modality"] == CT:
study_instance_uid = row["Study Instance UID"]
series_instance_uid = row["Series Instance UID"]
```

Get the directory to the patient folder, all folders in the patient folder, folder with the CT scan dicom slices

```python
from pathlib import Path

patient_id = row["Patient ID"]
patient_folder = dataset_folder / patient_id
patient_scan_folders = list(Path(patient_folder).iterdir())
ct_folder = patient_folder / study_instance_uid / series_instance_uid
```

Transfer all the dicom slices into the patient folder

```python
import shutil

files = list(Path(ct_folder).iterdir())

for file in files:
    shutil.move(file, patient_folder)
```

Delete the other folders that were accessed

```python
for folder in patient_scan_folders:
    shutil.rmtree(folder)

# End of for loop
```
